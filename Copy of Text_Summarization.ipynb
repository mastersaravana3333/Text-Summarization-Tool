```python
from transformers import pipeline
```

```python
# Load pre-trained summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
input_text = """
Artificial Intelligence (AI) is one of the fastest growing technologies in the modern world.
It enables machines to learn from data, identify patterns, and make decisions with minimal
human intervention. AI systems are widely used in healthcare for disease diagnosis,
in education for personalized learning, in finance for fraud detection, and in automation
for improving productivity. With the rapid growth of data and computational power,
AI is transforming industries and shaping the future of technology.
"""
```

```python
summary = summarizer(
    input_text,
    max_length=50,
    min_length=20,
    do_sample=False
)

import textwrap
print("ORIGINAL TEXT:\n")
print(input_text)
print("SUMMARY:\n")
print(textwrap.fill(summary[0]['summary_text'], width=80))
```
